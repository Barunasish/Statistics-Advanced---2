```python
"""
Question 1: What is hypothesis testing in statistics?

Answer: In statistics, hypothesis testing is a method for evaluating assertions about a population using sample data. It entails developing a null hypothesis is a statement indicating no effect or difference and an alternative hypothesis a statement that contradicts the null hypothesis.
We analyse sample data to assess whether there is sufficient evidence to reject the null hypothesis in favour of the alternative.
The null hypothesis (H0) : states that there is no impact or difference. Such as "there is no difference in average height between men and women" .
Alternative Hypothesis (HA):A remark that goes against the null hypothesis. Such as "there is a difference in average height between men and women" .
"""
```

```python
"""
Question 2: What is the null hypothesis, and how does it differ from the alternative hypothesis?

Answer: The null hypothesis (H0) is a declaration that there is no impact or difference, and it is the default assumption that researchers attempt to disprove. The alternative hypothesis (Ha or H1) is a statement that contradicts the null hypothesis and suggests a meaningful effect or relationship.
In essence, researchers seek evidence to either reject the null hypothesis and support the alternative, or to fail to reject the null hypothesis,implying that there is insufficient evidence to support the alternative.
The key difference between the two hypothesis is that the null hypothesis (H₀) implies that there is no effect, difference, or relationship in the population, sometimes known as the "status quo" assumption.
The alternative hypothesis (H₁ or Ha) suggests that there is an effect, difference, or relationship.
Hypothesis testing evaluates if the evidence is sufficient to reject H₀.

"""
```

```python
"""
Question 3: Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.

Answer: In hypothesis testing, the significance level (denoted by α, alpha) is the likelihood of rejecting the null hypothesis when it is true.
It is essentially a threshold established by the researcher that specifies how much evidence is required to reject the null hypothesis.
If the p-value (the likelihood of witnessing the data under the null hypothesis) is less than the significance level, the null hypothesis is rejected.
The significance level serves as a standard for assessing statistical significance.
By comparing the p-value to the significance level, we may determine whether the observed results are due to chance or a genuine effect.
A lower significance threshold (e.g., 0.01) necessitates more evidence to reject the null hypothesis, making it more difficult to find statistically significant results while decreasing the likelihood of a Type I error.
A higher significance level (e.g., 0.10) necessitates less evidence, making it easier to uncover statistically significant results while raising the possibility of a Type I mistake.

"""
```

```python
"""
Question 4: What are Type I and Type II errors? Give examples of each?

Answer: Type I and Type II errors in hypothesis testing occur when making conclusions about a population from sample data. A Type I mistake (false positive) happens when a valid null hypothesis is incorrectly rejected.
A Type II error (false negative) happens when an incorrect null hypothesis is not rejected.
Examples for the same:
Type I Error (False Positive): A pharmaceutical company is testing a novel medicine. The null hypothesis assumes that the medicine has no effect. As a result, the study concludes that the medicine is effective (rejects the null hypothesis), even if it has no impact.
The researchers falsely identify a positive outcome (the medicine works), which is not true.
Type II Error: A medical test is intended to diagnose an illness. The null hypothesis holds that the patient does not have the condition. As a result, the test inaccurately reveals that the patient is healthy (fails to reject the null hypothesis) whereas in reality they have the condition.
The test does not detect a true positive case (the patient has the disease).

"""
```

```python
"""

Question 5: What is the difference between a Z-test and a T-test? Explain when to use each?

Answer: The fundamental distinction between a Z-test and a T-test is the assumption of the population standard deviation and sample size. Z-tests are used when the population standard deviation is known and the sample size is big.
T-tests are used when the population standard deviation is unknown and the sample size is small.
The Z-test is used to compare a sample mean to a population mean with known standard deviation (e.g., in quality control or large-scale surveys).
T-test is used when comparing the means of two samples or a sample mean to the population mean and the population standard deviation is unknown (e.g., in medical research, educational studies, or social science research).

"""
```

```python
"""
Question 6: Write a Python program to generate a binomial distribution with n=10 and p=0.5, then plot its histogram

"""

import numpy as np
import matplotlib.pyplot as plt

# Parameters for the binomial distribution
n = 10  # number of trials
p = 0.5  # probability of success

# Generate random numbers following a binomial distribution
num_samples = 1000  # number of samples to generate
data = np.random.binomial(n, p, num_samples)

# Plot the histogram
plt.hist(data, bins=np.arange(-0.5, n + 1.5, 1), density=True, alpha=0.7, color='orange', edgecolor='purple')
plt.title('Binomial Distribution Histogram (n=10, p=0.5)')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.xticks(range(n + 1))
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
"""
Question 7: Implement hypothesis testing using Z-statistics for a sample dataset in Python. Show the Python code and interpret the results.
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

"""
import numpy as np
import scipy.stats as stats

# Sample data
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
                50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
                50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
                50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Parameters
mu_0 = 50  # Null hypothesis mean
alpha = 0.05  # Significance level

# Sample statistics
n = len(sample_data)
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)  # Sample standard deviation

# Z-statistic calculation
z_statistic = (sample_mean - mu_0) / (sample_std / np.sqrt(n))

# Critical value for two-tailed test
z_critical = stats.norm.ppf(1 - alpha / 2)

# p-value calculation
p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))

# Output results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Sample Standard Deviation: {sample_std:.2f}")
print(f"Z-statistic: {z_statistic:.2f}")
print(f"Critical Value: ±{z_critical:.2f}")
print(f"P-value: {p_value:.4f}")

# Decision
if abs(z_statistic) > z_critical:
    print("Reject the null hypothesis (H0).")
else:
    print("Fail to reject the null hypothesis (H0).")
```

```python
"""
Question 8: Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.

"""
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Parameters for the normal distribution
mu = 50  # mean
sigma = 5  # standard deviation
n = 100  # sample size

# Simulate data from a normal distribution
data = np.random.normal(mu, sigma, n)

# Calculate the sample mean and standard error
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # Sample standard deviation
standard_error = sample_std / np.sqrt(n)

# Calculate the 95% confidence interval
confidence_level = 0.95
z_critical = stats.norm.ppf((1 + confidence_level) / 2)  # Z critical value for 95% CI
margin_of_error = z_critical * standard_error

confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

# Output results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Sample Standard Deviation: {sample_std:.2f}")
print(f"95% Confidence Interval: {confidence_interval}")

# Plot the data
plt.figure(figsize=(10, 6))
plt.hist(data, bins=30, alpha=0.7, color='purple', edgecolor='orange', density=True)
plt.axvline(sample_mean, color='red', linestyle='dashed', linewidth=1, label='Sample Mean')
plt.axvline(confidence_interval[0], color='green', linestyle='dashed', linewidth=1, label='95% CI Lower Bound')
plt.axvline(confidence_interval[1], color='green', linestyle='dashed', linewidth=1, label='95% CI Upper Bound')
plt.title('Histogram of Simulated Data from Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.grid()
plt.show()
```

```python
"""

Question 9: Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram. Explain what the Z-scores represent
in terms of standard deviations from the mean.
"""

import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    # Convert data to a numpy array if it's not already
    data = np.array(data)

    # Calculate the mean and standard deviation
    mean = np.mean(data)
    std_dev = np.std(data, ddof=0)  # Population standard deviation

    # Calculate Z-scores
    z_scores = (data - mean) / std_dev

    # Plot the histogram of Z-scores
    plt.figure(figsize=(10, 6))
    plt.hist(z_scores, bins=15, alpha=0.7, color='blue', edgecolor='black', density=True)
    plt.title('Histogram of Z-scores')
    plt.xlabel('Z-score')
    plt.ylabel('Density')
    plt.grid()
    plt.axvline(0, color='red', linestyle='dashed', linewidth=1, label='Mean (Z=0)')
    plt.legend()
    plt.show()

    return z_scores

# Example usage
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
                50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
                50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
                50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

z_scores = calculate_z_scores(sample_data)
print("Z-scores:", z_scores)
```
